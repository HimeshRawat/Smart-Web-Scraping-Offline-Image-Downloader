```python
import os
import requests

# Base URL
base_url = "https://books.toscrape.com/catalogue/page-{}.html"

# Headers for browser mimic
headers = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/120.0.0.0 Safari/537.36"
    )
}

if not os.path.exists("htmls"):
    os.makedirs("htmls")

for i in range(1, 51):
    url = base_url.format(i)
    try:
        response = requests.get(url, headers=headers, timeout=10)
        response.raise_for_status()  
        with open(f"htmls/page{i}.html", "w", encoding="utf-8") as f:
            f.write(response.text)
        print(f"Downloaded page {i} successfully")
    except Exception as e:
        print(f"Failed to download page {i}: {e}")
```

```python
import os
from bs4 import BeautifulSoup
import requests

start_page = 1  
end_page = 2    

# Images folder banana
os.makedirs("images", exist_ok=True)

for page_num in range(start_page, end_page + 1):
    html_file = f"htmls/page{page_num}.html"
    
    
    if not os.path.exists(html_file):
        print(f" File {html_file} nahi mili, skip kar rahe hain...")
        continue

   
    with open(html_file, "r", encoding="utf-8") as f:
        soup = BeautifulSoup(f, "html.parser")

    
    img_tags = soup.find_all("img")

    print(f" Page {page_num} me {len(img_tags)} images mili.")

    for idx, img in enumerate(img_tags, start=1):
        img_url = img.get("src")

       
        if img_url.startswith("../"):
            img_url = "https://books.toscrape.com/" + img_url.replace("../", "")

        try:
            img_data = requests.get(img_url).content
            with open(f"images/page{page_num}_img{idx}.jpg", "wb") as f:
                f.write(img_data)
            print(f" Downloaded page{page_num}_img{idx}.jpg")
        except Exception as e:
            print(f" Error downloading {img_url}: {e}")
```
